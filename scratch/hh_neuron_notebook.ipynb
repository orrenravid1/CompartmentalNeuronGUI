```python
from neuron import h
import compneurovis as cnv

soma = h.Section(name='soma')
soma.L = 12.6157
soma.diam = 12.6157
soma.nseg = 1
soma.insert('hh')

stim = h.IClamp(soma(0.5))
stim.delay = 0.0
stim.dur = 1e9
stim.amp = 0.1

h.dt = 0.025
h.celsius = 6.3
h.finitialize(-65.0)

sim = cnv.source(cnv.neuron.attach(sections=[soma]))
sim.control(
    'clamp_amp',
    label='IClamp amplitude (nA)',
    get=lambda: float(stim.amp),
    set=lambda v: setattr(stim, 'amp', float(v)),
    min=-0.2,
    max=0.5,
)
```
